```python
df = pd.read_csv('car_price_prediction.csv')
df.head()
```

```python
df['horsepower'].plot.hist(bins=20, edgecolor='black')

plt.title('Распределение количества лошадиных сил')
plt.xlabel('Количество лошадиных сил')
plt.ylabel('Сколько раз встречается в датасете')

plt.show()
```

```python
df.plot.scatter(x='price', y='horsepower')

plt.title('Связь между стоимостью автомобиля и кол-вом ЛС')
plt.xlabel('Стоимость автомобиля')
plt.ylabel('Количество лошадиных сил')

plt.show()
```

С увеличением количества лошадиных сил увеличивается и стоимость автомобиля. Возрастающий тренд

Самостоятельная работа | Seaborn

```python
import seaborn as sns

# Создание ящика с усами для столбца price
sns.boxplot(x=df['price'])

# Добавление заголовка и подписей осей
plt.title('Ящик с усами для поиска выбросов в данных о стоимости автомобилей')
plt.xlabel('Стоимость автомобиля')

# Отображение графика
plt.show()
```

Как вы думаете, зачем нужна медиана, и почему на графике нет среднего значения?

Среднее значение же напротив, чувствительно к выбросам, поэтому оно может быть ненадежной мерой центральной тенденции.
