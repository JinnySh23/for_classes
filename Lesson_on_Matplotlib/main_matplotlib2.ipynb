```python
import pandas as pd
import matplotlib.pyplot as plt

# открыть файл с помощью сгенерированного пути
df = pd.read_csv('CarPrice_Assignment.csv')

# посмотреть 5 строк
df.head()
```

```python
plt.hist(df['price'])
plt.show()
```

```python
plt.hist(df['price'], bins=50)
plt.show()
```

```python
plt.hist(df['price'], bins=50)
plt.title('Распределение цен на автомабили')
plt.xlabel('Стоимость автомобилей')
plt.ylabel('Количество автомобилей данной стоимости')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x='horsepower', y='price', data=df) 
plt.title('Зависимость / влияние количества лошадиных сил в автомобиле на стоимость автомобиля')
plt.show()
```

```python
sns.scatterplot(x='enginesize', y='price', data=df) # 
plt.title('Зависимость / влияние объема двигателя на стоимость автомобиля')
plt.show()
```

```python
df_count = df[['enginesize', 'horsepower', 'price', 'peakrpm']] # оставляем только столбцы с числовыми значениями, которые нам интересны
df_count
```

```python
corr_matrix = df_count.corr() # считаем силу связи, корреляцию между числовыми параметрами
corr_matrix
```

```python
sns.heatmap(corr_matrix)
plt.show()
```

```python
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.boxplot(y='price', data=df) # создаем ящик с усами вертикальный
plt.title('Информация о ценах на автомобили')
plt.show()
```

```python
sns.boxplot(x='price', data=df) # создаем ящик с усами горизонтальный
plt.title('Информация о ценах на автомобили')
plt.show()
```
